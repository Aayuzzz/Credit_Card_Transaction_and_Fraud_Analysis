```python
#Handling missing zipcodes

df["zipcode_available"] = df["merch_zipcode"].notna()
df["merch_zipcode"] = df["merch_zipcode"].fillna("UNKNOWN")
```

```python
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
```

```python
#Extract transaction hour
df["transaction_hour"]=df["trans_date_trans_time"].dt.hour

# Extract day name
df["transaction_day"] = df["trans_date_trans_time"].dt.day_name()
```

```python
#Loading data into MySQL DATABASE

from sqlalchemy import create_engine
#MySQL connection
username="root" #default username
password="sql123" #your sql password
host="localhost" #if run locally
port="3306" #default MySQL port
database="credit_card_transactions" #The database you created in MySQL

engine=create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

#Write DataFrame to MySQL
table_name="transaction"
df.to_sql(table_name,engine,if_exists="replace",index=False)

#Read back sample
pd.read_sql("SELECT * FROM transaction LIMIT 5;",engine)
```
